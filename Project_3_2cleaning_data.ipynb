```python
from google.colab import files
import zipfile
import os

uploaded = files.upload()

for fn in uploaded.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(uploaded[fn])))

  # Unzip the uploaded file if it's a zip file
  if fn.endswith(".zip"):
    with zipfile.ZipFile(fn, 'r') as zip_ref:
        zip_ref.extractall(os.path.splitext(fn)[0])
    print(f"Unzipped {fn} to {os.path.splitext(fn)[0]}/")
```

```python
import pandas as pd
import os

# List files in the unzipped directory
unzipped_dir = os.path.splitext(list(uploaded.keys())[0])[0]
print(f"Files in {unzipped_dir}:")
for file in os.listdir(unzipped_dir):
    print(file)

# Assuming the main data is in a CSV file, load it into a DataFrame
# We'll look for a CSV file in the unzipped directory.
csv_files = [f for f in os.listdir(unzipped_dir) if f.endswith('.csv')]

if csv_files:
    # Assuming the first CSV file is the main data file
    csv_file_path = os.path.join(unzipped_dir, csv_files[0])
    df = pd.read_csv(csv_file_path)
    print(f"\nLoaded data from {csv_file_path}")
    display(df.head())
else:
    print("\nNo CSV files found in the unzipped directory.")
    df = None # Or handle appropriately if no CSV is expected
```

```python
import glob
import json
import os

# Load all CSV files into a dictionary of DataFrames, trying different encodings
csv_files_paths = glob.glob(os.path.join(unzipped_dir, '*.csv'))
dfs_csv = {}
for csv_file in csv_files_paths:
    country_code = os.path.basename(csv_file).split('videos.csv')[0]
    try:
        dfs_csv[country_code] = pd.read_csv(csv_file, encoding='utf-8')
        print(f"Successfully loaded {csv_file} with utf-8 encoding.")
    except UnicodeDecodeError:
        try:
            dfs_csv[country_code] = pd.read_csv(csv_file, encoding='latin1')
            print(f"Successfully loaded {csv_file} with latin1 encoding.")
        except Exception as e:
            print(f"Could not load {csv_file} with utf-8 or latin1 encoding: {e}")
            # Optionally, try another encoding or skip the file
            continue


# Load all JSON files into a dictionary
json_files_paths = glob.glob(os.path.join(unzipped_dir, '*.json'))
json_data = {}
for json_file in json_files_paths:
    country_code = os.path.basename(json_file).split('_category_id.json')[0]
    try:
        with open(json_file, 'r', encoding='utf-8') as f:
            json_data[country_code] = json.load(f)
        print(f"Successfully loaded {json_file} with utf-8 encoding.")
    except UnicodeDecodeError:
         try:
            with open(json_file, 'r', encoding='latin1') as f:
                json_data[country_code] = json.load(f)
            print(f"Successfully loaded {json_file} with latin1 encoding.")
         except Exception as e:
             print(f"Could not load {json_file} with utf-8 or latin1 encoding: {e}")
             continue
    except Exception as e:
        print(f"Could not load {json_file}: {e}")
        continue


# Display head of the first loaded CSV dataframe as an example
if dfs_csv:
    first_csv_key = list(dfs_csv.keys())[0]
    print(f"\nFirst loaded CSV DataFrame ({first_csv_key}):")
    display(dfs_csv[first_csv_key].head())

# Display keys of loaded json data
if json_data:
    print("\nKeys of loaded JSON data:")
    print(json_data.keys())
```

```python
for country, df in dfs_csv.items():
    print(f"\n--- Missing values in {country} before handling ---")
    missing_before = df.isnull().sum()
    print(missing_before[missing_before > 0])

    # Strategy: Drop rows with any missing values.
    # This is a simple approach and might not be suitable for all datasets,
    # but given the potentially large number of rows and relatively few columns,
    # dropping rows might be acceptable if missing values are not widespread.
    # A more nuanced approach would involve examining the percentage of missing values
    # per column and potentially imputing for some columns while dropping others.
    initial_rows = df.shape[0]
    dfs_csv[country] = df.dropna()
    rows_after_drop = dfs_csv[country].shape[0]
    print(f"Dropped {initial_rows - rows_after_drop} rows with missing values in {country}.")


    print(f"\n--- Missing values in {country} after handling ---")
    missing_after = dfs_csv[country].isnull().sum()
    print(missing_after[missing_after > 0])
```

```python
for country, df in dfs_csv.items():
    print(f"\n--- Duplicate rows in {country} before handling ---")
    initial_duplicates = df.duplicated().sum()
    print(f"Number of duplicate rows: {initial_duplicates}")

    # Remove duplicate rows
    dfs_csv[country].drop_duplicates(inplace=True)

    print(f"\n--- Duplicate rows in {country} after handling ---")
    duplicates_after_drop = dfs_csv[country].duplicated().sum()
    print(f"Number of duplicate rows: {duplicates_after_drop}")
```

```python
for country, df in dfs_csv.items():
    print(f"\n--- Processing {country} DataFrame ---")

    # Convert 'trending_date' to datetime
    try:
        dfs_csv[country]['trending_date'] = pd.to_datetime(df['trending_date'], format='%y.%d.%m')
        print(f"Converted 'trending_date' to datetime for {country}.")
    except Exception as e:
        print(f"Error converting 'trending_date' for {country}: {e}")

    # Convert 'publish_time' to datetime
    try:
        dfs_csv[country]['publish_time'] = pd.to_datetime(df['publish_time'])
        print(f"Converted 'publish_time' to datetime for {country}.")
    except Exception as e:
        print(f"Error converting 'publish_time' for {country}: {e}")

    # Convert numerical columns to numeric, coercing errors
    numerical_cols = ['views', 'likes', 'dislikes', 'comment_count']
    for col in numerical_cols:
        try:
            dfs_csv[country][col] = pd.to_numeric(df[col], errors='coerce')
            # Handle potential NaNs introduced by coercion if necessary (e.g., fill with 0 or median)
            # For this task, we'll just convert and note if NaNs are created.
            if dfs_csv[country][col].isnull().sum() > 0:
                 print(f"Warning: NaNs introduced in '{col}' column during conversion for {country}.")
            print(f"Converted '{col}' to numeric for {country}.")
        except Exception as e:
            print(f"Error converting '{col}' to numeric for {country}: {e}")

    print(f"Finished processing {country} DataFrame.")

# Display info for one dataframe to verify changes
if dfs_csv:
    first_key = list(dfs_csv.keys())[0]
    print(f"\n--- Info for {first_key} DataFrame after formatting ---")
    dfs_csv[first_key].info()
```

```python
import numpy as np

numerical_cols = ['views', 'likes', 'dislikes', 'comment_count']

for country, df in dfs_csv.items():
    print(f"\n--- Handling outliers for {country} ---")

    for col in numerical_cols:
        if col in df.columns:
            # Calculate Q1, Q3, and IQR
            Q1 = df[col].quantile(0.25)
            Q3 = df[col].quantile(0.75)
            IQR = Q3 - Q1

            # Define outlier bounds
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR

            print(f"  Column '{col}': Q1={Q1:.2f}, Q3={Q3:.2f}, IQR={IQR:.2f}, Lower Bound={lower_bound:.2f}, Upper Bound={upper_bound:.2f}")

            # Identify outliers
            outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
            print(f"  Identified {len(outliers)} outliers in '{col}'.")

            # Cap outliers
            dfs_csv[country][col] = np.where(dfs_csv[country][col] > upper_bound, upper_bound, dfs_csv[country][col])
            dfs_csv[country][col] = np.where(dfs_csv[country][col] < lower_bound, lower_bound, dfs_csv[country][col])
            print(f"  Capped outliers in '{col}'.")

            # Optional: Verify capping by checking min/max after capping
            # print(f"  After capping, min {col}: {dfs_csv[country][col].min():.2f}, max {col}: {dfs_csv[country][col].max():.2f}")

    print(f"Finished outlier handling for {country}.")

# Display summary statistics for one dataframe to observe the effect of capping
if dfs_csv:
    first_key = list(dfs_csv.keys())[0]
    print(f"\n--- Summary statistics for {first_key} DataFrame after outlier capping ---")
    display(dfs_csv[first_key][numerical_cols].describe())
```

```python
numerical_cols = ['views', 'likes', 'dislikes', 'comment_count']

for country, df in dfs_csv.items():
    print(f"\n--- Data Integrity Check for {country} ---")

    # 2. Print the shape
    print(f"Shape of DataFrame: {df.shape}")

    # 3. Check for remaining missing values
    print("\nRemaining missing values:")
    missing_after = df.isnull().sum()
    print(missing_after[missing_after > 0])

    # 4. Check for remaining duplicate rows
    print("\nRemaining duplicate rows:")
    duplicates_after = df.duplicated().sum()
    print(f"Number of duplicate rows: {duplicates_after}")

    # 5. Print data types
    print("\nData types of columns:")
    df.info()

    # 6. Print descriptive statistics for numerical columns
    print("\nDescriptive statistics for numerical columns:")
    if all(col in df.columns for col in numerical_cols):
        display(df[numerical_cols].describe())
    else:
        print("Numerical columns not found in DataFrame.")

    print(f"--- Finished Data Integrity Check for {country} ---")
```
