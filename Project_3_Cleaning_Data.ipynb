```python
from google.colab import files
import zipfile
import io

uploaded = files.upload()

for fn in uploaded.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(uploaded[fn])))

  if fn.endswith('.zip'):
    with zipfile.ZipFile(io.BytesIO(uploaded[fn]), 'r') as zip_ref:
        zip_ref.extractall('.')
    print(f"'{fn}' extracted.")
  else:
    print(f"'{fn}' is not a zip file, skipping extraction.")
```

```python
import pandas as pd

df = pd.read_csv('AB_NYC_2019.csv')
df.head()
```

```python
missing_values = df.isnull().sum()
print(missing_values)
```

```python
df['name'].fillna('Unknown Name', inplace=True)
df['host_name'].fillna('Unknown Host', inplace=True)
df['last_review'].fillna('No Review', inplace=True)
df['reviews_per_month'].fillna(0, inplace=True)
```

```python
missing_values_after_handling = df.isnull().sum()
print(missing_values_after_handling)
```

```python
num_duplicates = df.duplicated().sum()
print(f"Number of duplicate rows before removal: {num_duplicates}")
```

```python
df.info()
```

```python
df['last_review'] = pd.to_datetime(df['last_review'], errors='coerce')
df.info()
```

```python
import numpy as np

numerical_cols = ['price', 'minimum_nights', 'number_of_reviews', 'calculated_host_listings_count', 'availability_365']

outlier_bounds = {}
for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outlier_bounds[col] = (lower_bound, upper_bound)

    print(f"Column: {col}")
    print(f"  Q1: {Q1}, Q3: {Q3}, IQR: {IQR}")
    print(f"  Lower Bound: {lower_bound}, Upper Bound: {upper_bound}")
```

```python
for col, (lower, upper) in outlier_bounds.items():
    df[col] = np.where(df[col] > upper, upper, df[col])
    # We are only capping the upper bound based on the calculated IQR.
    # No action is needed for lower bound based on domain knowledge (e.g. price cannot be negative).

```

```python
print("Maximum values after capping outliers:")
for col in numerical_cols:
    print(f"  {col}: {df[col].max()}")
```
