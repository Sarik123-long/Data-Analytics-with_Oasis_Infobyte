```python
import pandas as pd

file_name = 'PlayStoreGameAppInfoReview.json'
df = pd.read_json(file_name)

display(df.head())
df.info()
```

```python
df_transposed = df.T
reviews_list = df_transposed['reviews'].tolist()

flat_reviews_list = [item for sublist in reviews_list for item in sublist]

reviews_df = pd.DataFrame(flat_reviews_list)

display(reviews_df.head())
reviews_df.info()
```

```python
# Display the first few rows
display(reviews_df.head())

# Print column names and data types
reviews_df.info()

# Check for missing values
display(reviews_df.isnull().sum())

# Get descriptive statistics for 'score'
display(reviews_df['score'].describe())

# Print unique values and their counts for 'score'
display(reviews_df['score'].value_counts())

# Display some sample reviews
print("Sample reviews from 'content' column:")
for i in range(5):
    if i < len(reviews_df):
        print(f"- {reviews_df['content'].iloc[i]}")
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

nltk.download('punkt', quiet=True)
nltk.download('stopwords', quiet=True)
nltk.download('wordnet', quiet=True)

# 1. Handle missing values
reviews_df['content'] = reviews_df['content'].fillna('')

# 2. Convert to lowercase and 3. Remove punctuation
reviews_df['cleaned_content'] = reviews_df['content'].str.lower().str.replace(r'[^\w\s]', '', regex=True)

# 4. Tokenization
reviews_df['cleaned_content'] = reviews_df['cleaned_content'].apply(lambda x: x.split())

# 5. Remove stop words
stop_words = set(stopwords.words('english'))
reviews_df['cleaned_content'] = reviews_df['cleaned_content'].apply(lambda x: [word for word in x if word not in stop_words])

# 6. Apply lemmatization
lemmatizer = WordNetLemmatizer()
reviews_df['cleaned_content'] = reviews_df['cleaned_content'].apply(lambda x: [lemmatizer.lemmatize(word) for word in x])

# Display the first few rows with the new cleaned column
display(reviews_df[['content', 'cleaned_content']].head())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Join the list of tokens back into strings
reviews_df['cleaned_content_str'] = reviews_df['cleaned_content'].apply(lambda x: ' '.join(x))

# Initialize TfidfVectorizer
# Consider max_features to limit vocabulary size
tfidf_vectorizer = TfidfVectorizer(max_features=5000)

# Apply TF-IDF to the cleaned text
tfidf_features = tfidf_vectorizer.fit_transform(reviews_df['cleaned_content_str'])

# Display the shape of the resulting feature matrix
print("Shape of TF-IDF feature matrix:", tfidf_features.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report

# 1. Create target variable
reviews_df['sentiment'] = reviews_df['score'].apply(lambda score: 0 if score in [1, 2] else (1 if score in [4, 5] else None))

# Drop rows with neutral sentiment (score 3)
reviews_df.dropna(subset=['sentiment'], inplace=True)
# Get the indices of the rows to keep
indices_to_keep = reviews_df.index

# Filter tfidf_features using the indices
tfidf_features_filtered = tfidf_features[indices_to_keep]


# Convert sentiment to integer type
reviews_df['sentiment'] = reviews_df['sentiment'].astype(int)

# 2. Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(tfidf_features_filtered, reviews_df['sentiment'], test_size=0.25, random_state=42)

# 3. Choose a classification model (Multinomial Naive Bayes)
model = MultinomialNB()

# 4. Train the model
model.fit(X_train, y_train)

# Evaluate the model (optional, but good practice)
y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import accuracy_score, classification_report

# 1. Make predictions on the test set
y_pred = model.predict(X_test)

# 2. Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)

# 3. Generate classification report
class_report = classification_report(y_test, y_pred)

# 4. Print accuracy and classification report
print(f"Accuracy: {accuracy}")
print("Classification Report:\n", class_report)
```

```python
# 1. Use the trained model to predict sentiment on X_test
y_pred = model.predict(X_test)

# 2. Add the predicted sentiment labels to the reviews_df DataFrame
# Create a DataFrame from X_test indices and y_pred
predicted_sentiment_df = pd.DataFrame({'predicted_sentiment': y_pred}, index=y_test.index)

# Merge with the original reviews_df on index
# Use a left merge to keep all rows in reviews_df that were in the test set
reviews_df = reviews_df.merge(predicted_sentiment_df, left_index=True, right_index=True, how='left')

# Fill NaN values in 'predicted_sentiment' for rows not in the test set (if any)
# For this task, we are focusing on the test set predictions, so we might just need to filter reviews_df
# to include only rows that were in the test set indices.
reviews_df_test = reviews_df.loc[y_test.index].copy()
reviews_df_test['predicted_sentiment'] = y_pred


# 3. Calculate and print the value counts for the 'predicted_sentiment' column
print("Distribution of Predicted Sentiments:")
display(reviews_df_test['predicted_sentiment'].value_counts())

# 4. Access the model's coefficients (feature_log_prob_)
log_probs = model.feature_log_prob_

# 5. Get feature names
feature_names = tfidf_vectorizer.get_feature_names_out()

# 6. For each sentiment class (0 and 1), sort the feature indices by their log probabilities
# Class 0 (Negative)
neg_class_prob_sorted_indices = log_probs[0, :].argsort()[::-1]
# Class 1 (Positive)
pos_class_prob_sorted_indices = log_probs[1, :].argsort()[::-1]

# 7. Print the top N feature names for each sentiment class
top_n = 20
print(f"\nTop {top_n} terms for Negative Sentiment (Class 0):")
display([feature_names[i] for i in neg_class_prob_sorted_indices[:top_n]])

print(f"\nTop {top_n} terms for Positive Sentiment (Class 1):")
display([feature_names[i] for i in pos_class_prob_sorted_indices[:top_n]])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud

# 1. Create a count plot of the predicted_sentiment column
plt.figure(figsize=(8, 6))
sns.countplot(x='predicted_sentiment', data=reviews_df_test)
plt.title('Distribution of Predicted Sentiments')
plt.xlabel('Predicted Sentiment')
plt.ylabel('Number of Reviews')
plt.xticks(ticks=[0, 1], labels=['Negative (0)', 'Positive (1)'])
plt.show()

# Prepare text for word clouds
# Join the cleaned content for each sentiment class
negative_text = " ".join(reviews_df_test[reviews_df_test['predicted_sentiment'] == 0]['cleaned_content_str'].tolist())
positive_text = " ".join(reviews_df_test[reviews_df_test['predicted_sentiment'] == 1]['cleaned_content_str'].tolist())

# Generate word cloud for negative sentiment
wordcloud_neg = WordCloud(width=800, height=400, background_color='white').generate(negative_text)

# Generate word cloud for positive sentiment
wordcloud_pos = WordCloud(width=800, height=400, background_color='white').generate(positive_text)

# Display the word clouds
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(wordcloud_neg, interpolation='bilinear')
plt.title('Word Cloud for Negative Sentiment')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(wordcloud_pos, interpolation='bilinear')
plt.title('Word Cloud for Positive Sentiment')
plt.axis('off')

plt.show()
```
